# 1. Импортируйте набор данных Fashion-MNIST:

```python
# 1. Import the Fashion-MNIST dataset:

from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import fashion_mnist
```

# 2. Создайте обучающую и тестовую выборки (тензоры, содержащие входные признаки и метки) с помощью метода load_data():

```python
# 2. Create training and test sets (tensors containing input features and labels) using the load_data() method:

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# Checking the size and the shape

x_train[0]

# Это пиксели, которые представляет каждое изображение
```

# 3. Для удобства создайте текстовые метки классов (на русском или английском языке)

```python
# 3. For convenience, create text class labels (in Russian or English):

classes = [
    'футболка', 'брюки', 'свитер', 'платье', 'пальто',
    'туфли', 'рубашка', 'кроссовки', 'сумка', 'ботинки'
]
```

# 4. Выведите любые 50 изображений из обучающей выборки вместе с соответствующими метками классов:

```python
# 4. Output any 50 images from the training set along with the corresponding class labels:

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

n = 100

for i in range(n, n + 50):
    plt.subplot(5, 10, i - n + 1)

    plt.xticks([])

    plt.yticks([])

    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)

    plt.xlabel(classes[y_train[i]])
```

# 5. Выведите размерность массива x_train, содержащего входные признаки обучающего набора. Преобразуйте его в двумерный массив размерности (60000, 784).

```python
# 5. Output the dimension of the array x_train containing the input features of the training set. Convert it to a two-dimensional array of dimension (60000, 784).

x_train = x_train.reshape(60000, 784)
print(x_train.shape)
```

```python
x_train[0]
```

# 6. Выполните масштабирование входных данных:

```python
# 6. Scale the input data:

x_train = x_train / 255

# Оно делится на 255, поскольку интенсивность пикселей представлена в диапазоне от 0 до 255.
```

# 7. Выведите метку любого объекта из y_train (должна быть цифра 0т 0 до 9).

```python
# 7. Print the label of any object from y_train (should be the number 0m 0 to 9).

print(y_train[0])
print(classes[y_train[0]])
```

# 8. Выполните бинарное кодирование меток классов:

```python
# 8. Binary encode the class labels:

from tensorflow.keras import utils

y_train = utils.to_categorical(y_train, 10)

print(y_train.shape)
```

```python
print(y_train[0])
```

# 9. Создайте полносвязную ИНС с одним скрытым слоем с 800 нейронами для решения задачи классификации. Скомпилируйте модель и выведите информацию о полученной архитектуре и количестве параметров модели. Проконтролируйте правильность архитектуры.

```python
# 9. Create a fully connected ANN with one hidden layer with 800 neurons to solve the classification problem. Compile the model and output information about the resulting architecture and the number of model parameters. Check the correctness of the architecture.

model = Sequential()
model.add(layers.Dense(800, activation='relu', input_shape=(784,)))
model.add(layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

# 10. Обучите созданную ИНС, задав размер батча, равным 200, и количество эпох 50. Выведите и проанализируйте результаты процесса обучения по эпохам.

```python
# 10. Train the created ANN by setting the batch size to 200 and the number of epochs to 50. Derive and analyze the results of the learning process by epoch.

history = model.fit(x_train, y_train, batch_size=200, epochs=50)

# batch_size = количество объектов из обучающей выборки, которые проходят по сети перед обновлением

# epochs = количество эпох обучения (полных проходов всех обучающих примеров через сеть)
```

# 11. Поэкспериментируйте с размером батчей, проследите, как это влияет на скорость и результаты обучения.

```python
# 11. Experiment with the batch size, see how it affects the speed and learning outcomes.

history_1 = model.fit(x_train, y_train, batch_size=100, epochs=50)
```
