```python
# Generate a dataset in the form of a cloud of 1500 points with three centers and variances of 1.0, 2.5, 0.5 (cluster_std parameter)
# and then apply K-means clustering to the data.

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
variances = [1.0, 2.5, 0.5]

# iterate over the variances
for variance in variances:
    # generate the dataset
    X, y = make_blobs(n_samples=1500, centers=3, cluster_std=variance)
    # plot the dataset
    plt.scatter(X[:, 0], X[:, 1], c='black')
    plt.title(f'original_{variance}.png')
    plt.show()
    # apply k-means clustering
    kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
    # plot the data
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
    # save the figure
    plt.title('reg_' + str(variance) + '.png')
    # close the figure
    plt.show()
```
